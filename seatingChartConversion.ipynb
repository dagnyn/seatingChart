```python
import pandas as pd

df = pd.read_excel('SeatingChart.xlsx')

guestDict = {}
allNames = []

# Create dictionary of guest names and tables
for col in df.columns:
    tableNum = col.split(" ")[1]
    # print("\nTable", tableNum)
    for n in df[col]:
      name = str(n)
      if name == "nan":
          continue
      if (name not in allNames):
        allNames.append(name)
        guestDict[name] = [tableNum]
      else:
         guestDict[name].append(tableNum)


# Sort alphabetically
allNames.sort()

# Format list of names
allNamesString = "let allNames = ["

for name in allNames:
  allNamesString = allNamesString + "\"" + name + "\", "

allNamesString = allNamesString + "]\n"
```

Write dictionary append strings to a file

```python
with open("chartHelper.txt", "w") as f:

  f.write(allNamesString)
  f.write("let seatingChart = new Map();\n")
  for name in guestDict:
    line = "seatingChart.set(\"" + name + "\", [" + ", ".join(guestDict[name]) + "]);\n"
    f.write(line)


f.close()
```
